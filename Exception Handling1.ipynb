Q1. Explain why we have to use the Exception class while creating a Custom Exception.
Note: Here Exception class refers to the base class for all the exceptions.

When creating a Custom Exception in Python, we usually inherit from the built-in Exception class because Exception is the base class for most runtime errors.

```python
class ageError(Exception):
  pass
try:
    age = -5
    if age < 0:
      raise ageError("age does not negative")

except ageError as e:


    print(e)

```

Q6. List down some best practices for exception handling in python.

1.Catch specific exceptions instead of using a generic except block.

```python
try:
    num = int("abc")
except ValueError:
    print("Invalid number")
```

2.Avoid using a bare except: clause.

```python
try:
    pass
except:
    print("Error")
```

3.Use multiple except blocks when handling different exceptions.

```python
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero")
except TypeError:
    print("Type error occurred")
```
