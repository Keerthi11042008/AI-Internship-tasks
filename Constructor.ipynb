9. Create a Python class called Rectangle with a constructor that initializes the width and height attributes. Provide a method to calculate the area of the rectangle.

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

r1 = Rectangle(5, 4)

print("Area =", r1.area())
```

10. How can you have multiple constructors in a Python class? Explain with an example.

Python does not support multiple constructors directly like some other programming languages. Since a class can have only one __init__() method, multiple constructors can be simulated using:

1.Default arguments
2.Variable-length arguments (*args, **kwargs)
3.Class methods (alternative constructors)

```python
class Student:
    def __init__(self, name=None, age=None):
        self.name = name
        self.age = age

    def display(self):
        print("Name:", self.name)
        print("Age:", self.age)


s1 = Student()
s1.display()

s2 = Student("Keerthika", 20)
s2.display()
```

**INHERITANCE:**

1. What is inheritance in Python? Explain its significance in object-oriented programming.

Inheritance is an OOP feature that allows one class (child/derived class) to acquire the properties and methods of another class (parent/base class).

Significance:
Promotes code reusability.
Reduces code duplication.
Makes programs easier to maintain.
Supports hierarchical classification.

```python
class Animal:
    def sound(self):
        print("Animal makes a sound")

class Dog(Animal):
    pass

d = Dog()
d.sound()
```

2. Differentiate between single inheritance and multiple inheritance in Python. Provide examples
for each.

**SINGLE INHERITANCE EXAMPLE:**

```python
class Parent:
    def show(self):
        print("Parent Class")

class Child(Parent):
    pass

c = Child()
c.show()
```

**MULTIPLE INHERITANCE EXAMPLE**

```python
class Father:
    def skill1(self):
        print("Driving")

class Mother:
    def skill2(self):
        print("Cooking")

class Child(Father, Mother):
    pass

c = Child()
c.skill1()
c.skill2()
```

3. Create a Python class called Vehicle with attributes color and speed. Then create a child class Car that inherits from Vehicle and adds a brand attribute.

```python
class Vehicle:
    def __init__(self, color, speed):
        self.color = color
        self.speed = speed

class Car(Vehicle):
    def __init__(self, color, speed, brand):
        super().__init__(color, speed)
        self.brand = brand

car1 = Car("Red", 120, "Toyota")

print("Color:", car1.color)
print("Speed:", car1.speed)
print("Brand:", car1.brand)
```

4. Explain the concept of method overriding in inheritance. Provide a practical example.

Method Overriding occurs when a child class provides its own implementation of a method that already exists in the parent class.

```python
class Animal:
    def sound(self):
        print("Animal makes a sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")

d = Dog()
d.sound()
```

5. How can you access the methods and attributes of a parent class from a child class in Python? Give an example.

1.Inheritance directly through the child object.
2.The super() function.

```python
class Parent:
    def __init__(self):
        self.name = "Python"

    def display(self):
        print("Parent Method")

class Child(Parent):
    def __init__(self):
        super().__init__()

    def show(self):
        print("Name:", self.name)
        super().display()

c = Child()
c.show()
```

4. Explain the concept of method overriding in polymorphism. Provide an example.

```python
class Animal:
    def sound(self):
        print("Animal makes a sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")

class Cat(Animal):
    def sound(self):
        print("Cat meows")

d = Dog()
c = Cat()

d.sound()
c.sound()
```

6. Create a Python class called `Animal` with a method `speak()`. Then, create child classes like
`Dog`, `Cat`, and `Bird`, each with their own `speak()` method. Demonstrate polymorphism by
calling the `speak()` method
on objects of different subclasses.

```python
class Animal:
    def speak(self):
        print("Animal makes a sound")

class Dog(Animal):
    def speak(self):
        print("Dog barks")

class Cat(Animal):
    def speak(self):
        print("Cat meows")

class Bird(Animal):
    def speak(self):
        print("Bird chirps")


dog = Dog()
cat = Cat()
bird = Bird()


animals = [dog, cat, bird]

for animal in animals:
    animal.speak()
```

8. Create a Python class hierarchy for a vehicle system (e.g., car, bicycle, boat) and implement

```python
class Vehicle:
    def start(self):
        print("Vehicle is starting")

class Car(Vehicle):
    def start(self):
        print("Car engine starts with a key")

class Bicycle(Vehicle):
    def start(self):
        print("Bicycle starts by pedaling")

class Boat(Vehicle):
    def start(self):
        print("Boat engine starts and moves on water")


car = Car()
bicycle = Bicycle()
boat = Boat()

vehicles = [car, bicycle, boat]

for vehicle in vehicles:
    vehicle.start()
```

15. Describe the concept of operator overloading in Python and how it relates to polymorphism.
Provide
examples using operators like `+` and `*`.

```python
class Student:
    def __init__(self, marks):
        self.marks = marks

    def __add__(self, other):
        return self.marks + other.marks

s1 = Student(80)
s2 = Student(90)

print(s1 + s2)
```

```python
class Number:
    def __init__(self, value):
        self.value = value

    def __mul__(self, other):
        return self.value * other.value

n1 = Number(5)
n2 = Number(6)

print(n1 * n2)
```

**ABSTRACTION:**

7. Discuss the concept of interface classes in Python and their role in achieving abstraction.

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):

    @abstractmethod
    def start(self):
        pass

class Car(Vehicle):
    def start(self):
        print("Car starts")

c = Car()
c.start()
```

Explain the significance of encapsulation in achieving abstraction. Provide examples.
Encapsulation

```python
class Student:
    def __init__(self):
        self.__marks = 80

    def get_marks(self):
        return self.__marks

s = Student()

print(s.get_marks())
```

0. What is the purpose of abstract methods, and how do they enforce abstraction in Python classes?

```python
from abc import ABC, abstractmethod

class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

class Circle(Shape):

    def area(self):
        return 3.14 * 5 * 5

c = Circle()
print(c.area())
```
