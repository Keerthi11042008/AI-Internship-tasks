20.
Find the longest word in a string.

```python
s="python programming is easy"
words=s.split()
longest=words[0]
for word in words:
  if len(word)>len(longest):
    longest=word
print("longest word:",longest)
```

21. Find the shortest word in a string.  

```python
s="python programming is easy"
words=s.split()
shortest=words[0]
for word in words:
    if len(word)<len(shortest):
        shortest=word
print("shortest word:",shortest)
```
